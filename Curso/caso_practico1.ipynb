```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style='darkgrid')
df = sns.load_dataset('tips')
df
```

```python
# Grafico de barras de propinas por sexo
sns.barplot(data=df, x='sex', y='tip', errorbar=None)
plt.title('Propinas promedio por sexo')
plt.show()
```

Primer analisis: Hombres pagan mas propina que mujeres

```python
sns.barplot(data=df, x='day', y='tip', errorbar=None)
plt.title('Promedio propinas por dia')
plt.show()
```

Segundo analisis: Los domingos se pagan mas propinas

```python
sns.barplot(data=df, x='time', y='tip', errorbar=None)
plt.title('Propinas por tiempo')
plt.show()
```

Tercer analisis: En las cenas se dan mas propinas.
