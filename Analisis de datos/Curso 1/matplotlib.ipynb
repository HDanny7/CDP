```python
import matplotlib.pyplot as plt
```

```python
# Ejemplo de una linea simple
x = [0, 1, 2, 3, 4 ,5]
y = [0, 1, 4, 9, 16, 25]
plt.plot(x, y)
plt.scatter(x, y)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Grafica sencilla')
plt.show()
```

```python
# Grafico de barras
categorias = ['A', 'B', 'C', 'D']
valores = [10, 20, 25, 30]
plt.bar(categorias, valores)
plt.title('Grafica de barra')
plt.xlabel('Ã‡ategorias')
plt.ylabel('Valores')
plt.show()
```
