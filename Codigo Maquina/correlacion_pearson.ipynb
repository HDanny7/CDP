# Correlacion de Pearson

Magnitud y direccion de una relacion lineal entre 2 variables

```python
# Importamos las librerías necesarias
import matplotlib.pyplot as plt
import pandas as pd

personas = pd.read_csv('personas.csv')
personas
```

# Realcion peso / altura

```python
plt.scatter(personas['altura'], personas['peso'])
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.show()
```

# Ingreso y horas trabajadas

```python
plt.scatter(personas['ingreso'], personas['horas_trabajadas'])
plt.xlabel('Ingreso ($)')
plt.ylabel('Horas Trabajadas')
plt.show()
```

# Ingresos y ausencias

```python
plt.scatter(personas['ingreso'], personas['ausencias'])
plt.xlabel('Ingresos ($)')
plt.ylabel('Ausencias')
plt.show()
```

# Ingreso y peso

```python
plt.scatter(personas['ingreso'], personas['peso'])
plt.xlabel('Ingresos ($)')
plt.ylabel('Peso (kg)')
plt.show()
```

# Materiz de correlacion

```python
matriz = personas.corr()
plt.matshow(matriz, cmap='coolwarm', aspect='auto', vmin= -1, vmax=1)
plt.colorbar()
plt.xticks(range(len(matriz.columns)), matriz.columns, rotation=90)
plt.yticks(range(len(matriz.columns)), matriz.columns)

for i in range(len(matriz.columns)):
    for j in range(len(matriz.columns)):
        plt.text(j, i, f'{matriz.iloc[i, j]:.2f}', ha='center', va='center', color='black')

plt.show()
```
